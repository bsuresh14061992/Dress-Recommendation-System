```python
import pandas as pd
from lxml import html
from selenium import webdriver as wb
from webdriver_manager.chrome import ChromeDriverManager
from time import sleep
webD=wb.Chrome(ChromeDriverManager().install())


lk=[]
ld=[]
lr=[]
ldp=[]
sav=[]
img=[]

for page_nb in range(1,3,1):
    webD.get("https://www.amazon.in/s?k=dresses+for+womens&i=apparel&page={}".format(page_nb))
    sleep(1)
    tree=html.fromstring(webD.page_source)




    for product_tree in tree.xpath("//div[contains(@data-cel-widget,'search_result_')]"):
        try:
            brand=product_tree.xpath(".//span[@class='a-size-base-plus a-color-base']/text()")
            lk.append(brand)
        except:
            brand=None
            lk.append(brand)
        try:    
            description=product_tree.xpath(".//span[@class='a-size-base-plus a-color-base a-text-normal']/text()")
            ld.append(description)
        except:
            description=None
            ld.append(description)
        try:
            rating=product_tree.xpath(".//span[@class='a-icon-alt']/text()")
            lr.append(rating)
        except:
            #print("ss")
            rating=None
            lr.append(rating)
        try:
            discount_price=product_tree.xpath(".//span[@class='a-price-whole']/text()")
            ldp.append(discount_price)
        except:
            discount_price=None
            ldp.append(discount_price)
        try:
            actual_price=product_tree.xpath(".//span[@aria-hidden='true']/text()")
            sav.append(actual_price)
        except:
            save_price=None
            sav.append(actual_price)
            
        

        print(brand,rating,discount_price,description,actual_price)#,save_price

    
#webD.close()  
```
